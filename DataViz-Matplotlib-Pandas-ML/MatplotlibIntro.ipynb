### Creating a simple chart
Create the data, then plot it into a chart and label the title and it's axis then show it

```python
# Creating two list to serve as the data point to be used in the chart
x=[1,2,3]
y=[1,4,9]

# to plot the data points into a chart use pyplots .plot(...) method
# first value-> x-axis, second value-> y-axis
plt.plot(x,y) 

#creating labels for the chart
plt.title("test plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

#printing the chart
plt.show()
```

```python
# to create a chart with more dimensions
x=[1,2,3]
y=[1,4,9] 
z=[10,7,3] #additional data points 

plt.plot(x,y) 
plt.plot(x,z)# plotting 2nd y-axis value, keep the same x value

plt.title("Multi-line plot test")
plt.xlabel("x-axis")
plt.ylabel("y & z -axis")

#To label lines in a legend
# the .legend method takes in a list as an input, and they are associated in order 
# in companion to the order that the lines were ploted (we ploted y fisrt and then z)
plt.legend(["this is y","this is z"]) 

plt.show()
```

### Accessing individual specific values form a df column

To access individual cell values from a df colum, you use the iloc method as pass in the index of the desired value

.iloc[specify desired column index here]

```python
#example:
# to access the 2nd indexed item in the "column_C" column you use this:
# <dataframe_name>.<column_name>.iloc[ <desired value index> ]
# remember indexig starts at 0 so the first item will be indexed 0 
# and the 2nd item indexed 1 and so on.
# Below accessing the third df column, and the second indexed item in that colum meaning the third item in that column

sample_data_df.column_c.iloc[2]
```

### Ploting the df data into a plot

```python
x= sample_data_df.column_a
y= sample_data_df.column_b

plt.plot(x,y)
plt.title("Sample df plot")
plt.xlabel("columb_b")
plt.ylabel("columb_a")
plt.show()
```

To add the third colum to the plot, repeat the previous step but use the .plot() method again to include the third column 

 Make sure to include the same column in the x axis.
 
You can also add the 'o' string following any datapoint in the plt.() method to format the plot, the 'o' flag specifies dots 

There are other flags that can be found in the Matplotlib documentation.

```python
x= sample_data_df.column_a
y= sample_data_df.column_b
z= sample_data_df.column_c #add this line

plt.plot(x,y,'x') # note we are using the 'x' flag, you can also use '.' and '_'
plt.plot(x,z, 'o') # add this line , note we use the 'o' flag for the z points


plt.title("Sample df plot 2 lines")
plt.xlabel("columb_a")
plt.ylabel("columb_b & columb_c")
plt.show()
```

## Real data visualization plot

We will use the countries.csv dataset file

```python
#loading countries.csv data into a dataframe
countries_df=pd.read_csv('countries.csv')

#printig a snippet of the dataframe to get an idea of what it looks like:
countries_df 
```

### Performing analysis
Comparing the population growth between the US and China

Step one: Isolating the data by filtering:

```python
#This will give us an indexe series of True where the country is labeled "United States"
# and False where it is not
countries_df.country == "United States"
```

```python
#When surrounding the above query with square brackets [] and adding the df name at the front
# this will actually list the values of the indexes that are labeled as True for the query
countries_df[countries_df.country == "United States"]
```

```python
# Filtering out US data and asssigning it to a variable
us = countries_df[countries_df.country == 'United States']

#printing it
us
```

```python
#Filtering out China data and assigning it to a variable
china = countries_df[countries_df.country == 'China']

#printing it
china
```

### Graphing the data

```python
#plotting us-year in the x axis and 
# plotting us population in the y axis
plt.plot(us.year,us.population)

#plotting china-year in the x axis(Note: plotting 2 different x-axis is fine since they both have the same years)
# and plotting china population in the y axis
plt.plot(china.year,china.population)

#adding labels to the plot
plt.title("China vs US population")
plt.ylabel("Population Growth")
plt.xlabel("year")

#addign a legend
plt.legend(["US","China"])
#displaying the graph
plt.show()

#NOTE: the y-axis label is denoted in scientific notation due to the large numbers
```

### Modifying the chart to change the y-axis to display in millions
not scientific notation

```python
# we only need to adjust the y axis on the 2 plots, and the y axis label
# using 10^6 to divide the populations by 1,000,000
# in python an exponent is denoted with **
plt.plot(us.year,us.population/ 10**6) # dividing population in 10**6 ->(1,000,000 a million)
plt.plot(china.year,china.population/ 10**6) # dividing population in 10**6 ->(1,000,000 a million)

plt.title("China vs US population")
plt.ylabel("Population Growth in Millions") #specify to let people know that the units are in millions now
plt.xlabel("year")

plt.legend(["US","China"])

plt.show()
```

## Growth in percentage rate

In contrast to the first year, to the first year in the dataset, 1952

```python
#plotting us-year in the x axis and 
# plotting us population in the y axis
plt.plot(us.year,us.population/us.population.iloc[0] * 100) # divide it by the 100% unit which would be the population in 1608 and I'd be in decimal form like .8 for 80% so we multiply at the end to convert to percentage form 

#plotting china-year in the x axis(Note: plotting 2 different x-axis is fine since they both have the same years)
# and plotting china population in the y axis
plt.plot(china.year,china.population/china.population.iloc[0] * 100)

#adding labels to the plot
plt.title("China vs US population")
plt.ylabel("Population growth rate compared to 1952")
plt.xlabel("year")

#addign a legend
plt.legend(["US","China"])
#displaying the graph
plt.show()
```

Example population growth rate percentage plot for 3 different countries since 1952:

Gaza, Israel, Taiwan

```python
import pandas as pd

df= pd.read_csv("countries.csv")

gaza=df[df.country == "West Bank and Gaza"]
israel = df[df.country == "Israel"]
taiwan= df[df.country == "Taiwan"]
from matplotlib import pyplot as plt

plt.plot(gaza.year,gaza.population/gaza.population.iloc[0]*100, '_')
plt.plot(israel.year,israel.population/israel.population.iloc[0]*100,'x')
plt.plot(taiwan.year,taiwan.population/taiwan.population.iloc[0]*100, '.')

plt.title("Gaza, Israel, and Taiwan")
plt.xlabel("year")
plt.ylabel("population growth %")

plt.legend(["GOAT GAZA","Israel","Taiwan"])

plt.show()
```
